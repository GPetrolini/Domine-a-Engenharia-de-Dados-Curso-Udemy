Conversão de datas

```python
df = pd.read_csv(r'C:\curso-eng-dados\Python-SGBDS\Origem de Dados\arquivos_csv\Clientes.csv', delimiter=',')
df
```

```python
print(df['created_at'].dtypes)
```

```python
# Converter a coluna 'create_at' para o formato data e hora
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
print(df['created_at'].dtypes)
```

```python
#filtrar os dados para o intervalo de anos entre 2018 e 2020
df_filtrado = df[(df['created_at'].dt.year >= 2018) & (df['created_at'].dt.year <= 2020)]
df_filtrado.head()
```

```python
# filtrar somente o ano de 2018 e o mes de janeiro
df_dt2 = df[(df['created_at'].dt.year >= 2018) & (df['created_at'].dt.month == 1)]

#imprimir o dataframe filtrado
df_dt2
```

Conversão formato brasileiro


```python
# Converter a coluna 'create_at' para o formato de data no padrao brasileiro
df['created_at'] = df['created_at'].dt.strftime('%d/%m/%Y')
```
